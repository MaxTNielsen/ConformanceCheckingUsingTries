```python
from logs import get_traces

from torch.nn.functional import one_hot

import torch as t
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
FILE_NAME = "logs/M1.xes"

traces_dict = {int(key[9:]):val for key,val in get_traces(filename=FILE_NAME).items()}

print(traces_dict)
```

```python
# build maps to map from and to activity labels and unique ids

labels = {val for values in traces_dict.values() for val in values}

labels_to_idx = {}
idx_to_labels = {}

for i,label in enumerate(labels):
    labels_to_idx[label] = i
    idx_to_labels[i] = label
```

```python
def make_data_set():
    data_set = []
    l = list(labels_to_idx.values())
    one_hot_enc = one_hot(t.arange(0, len(l)), num_classes=len(l))

    for trace in traces_dict.values():
        trace_ = []
        for activity in trace:
            activity_idx = labels_to_idx[activity]
            trace_.append(one_hot_enc[activity_idx].numpy())
        data_set.append(trace_)

    inputs = []

    for trace in data_set:
        input_trace = []
        input_trace.append(t.zeros(len(l)).numpy())
        input_trace = input_trace + trace[:-1]
        inputs.append(input_trace)

    # max_len = 0
    # for trace in data_set:
    #     if len(trace) > max_len:
    #         max_len = len(trace)

    # padded_data_set = []
    # for trace in data_set:
    #     for i in range(max_len-len(trace)):
    #         trace.append(t.zeros(len(l)).numpy())
    #     padded_data_set.append(trace)

    # return padded_data_set
    return inputs, data_set

outputs, targets = make_data_set()
```
